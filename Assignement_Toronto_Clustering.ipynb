```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from bs4 import BeautifulSoup
import requests
```

## scraping and transforming the table from wikipedia to dataframe

```python
# scraping wikipedia page
path = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(path, 'lxml')
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

```python
df_toronto = pd.read_html(tab)
df_toronto = df_toronto[0]
df_toronto.head()
```

## cleaning dataset 

```python
# drop the rows with the Neighborhood of 'Not assigned'
df = df_toronto[df_toronto['Neighbourhood']!= 'Not assigned']
df.head()
```

```python
# combine the rows
df_combine = df.groupby(['Postal Code'], sort=False).agg(','.join)
df_combine.reset_index(inplace=True)
```

```python
df_combine.head()
```

```python
df_combine.tail()
```

```python
df_combine.shape
```

## dealing with Latitudes and Longitudes

```python
# choose to use the given dataset insteading of using google geocoder
lat_lon = pd.read_csv('https://cocl.us/Geospatial_data')
lat_lon.head()
```

```python
# merge the two tables
df2 = pd.merge(df_combine, lat_lon, on='Postal Code')
```

```python
df2.head()
```

## clustering and plotting

```python
# drop all the rows which do not contain "toronto"
df3 = df2[df2['Borough'].str.contains('Toronto',regex=False)]
```

```python
df3.head()
```

```python
df3.tail()
```

```python
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)
```

```python
for lat,lng,borough,neighbourhood in zip(df3['Latitude'],df3['Longitude'],df3['Borough'],df3['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

## clustering Toronto with the method of KMeans

```python
from sklearn.cluster import KMeans
k=5
X_toronto = df3.drop(['Postal Code','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(X_toronto)
kmeans.labels_
df3.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
df3.head()
```

```python
df3.tail()
```

```python
import matplotlib
map_clustering = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = matplotlib.cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [matplotlib.colors.rgb2hex(i) for i in colors_array]

markers_colors = []
for lat, lon, neighbourhood, cluster in zip(df3['Latitude'], df3['Longitude'], df3['Neighbourhood'], df3['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clustering)
       
map_clustering
```
